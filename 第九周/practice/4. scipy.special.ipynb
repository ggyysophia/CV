```python
S = ss
print(S.round(3.5))
print(S.round(3.499))
print(S.comb(5,3))#从5个中任选3个的组合数
print(S.perm(5,3))#排列数
print(S.gamma(4))#gamma函数
print(S.beta(10,200))#beta函数
print(S.sinc(0))#sinc函数


```
